```python
import matplotlib
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Nombre del archivo CSV con múltiples hojas
archivo_csv = "/content/data/Base_2019.xlsx"

# Leer cada hoja del archivo CSV en un diccionario de DataFrames
diccionario_dataframes = pd.read_excel(archivo_csv, sheet_name=None)
```

### Selección preliminar de variables de interés


```python
columns_deleted = [
"Dia",
"Fecha",
"MES_PROCESADO",
"Oficina",
"GravedadCod",
"ClaseCodigo",
"ClaseNombre",
"ChoqueCodigo",
"ChoqueNombre",
"ObjetoFijoCodigo",
"ObjetoFijoNombre",
"OtraClase",
"NombreOtraClase",
"Direccion",
"TipoVia1",
"NumeroVia1",
"LetraVia1",
"CardinalVia1",
"TipoVia2",
"NumeroVia2",
"LetraVia2",
"CardinalVia2",
"Complemento",
"Municipio",
"Localidad",
"FechaOcurrencia",
"HORA_PROCESADA",
"Area",
"Sector",
"Zona",
"TipoDisenno",
"ZonaTransito",
"AreaTransito"
]

columns_to_consider = [
"idFormulario",
"GravedadNombre",
"Tipo_Accidente",
"Choque_Nombre",
"Latitud",
"Longitud",
"HoraOcurrencia",
"TipoTiempo",
"CON_BICICLETA",
"CON_CARGA",
"CON_EMBRIAGUEZ",
"CON_HUECOS",
"CON_MENORES",
"CON_MOTO",
"CON_PEATON",
"CON_PERSONA_MAYOR",
"CON_RUTAS",
"CON_TPI",
"CON_VELOCIDAD"
]

def Tipo_Accidente(row):
    return str(row['ClaseNombre']) + ' ' + str(row['NombreOtraClase'])

def Choque_Nombre(row):
    return str(row['ChoqueNombre']) + ' ' + str(row['ObjetoFijoNombre'])


ACCIDENTES = diccionario_dataframes['ACCIDENTES']
ACCIDENTES['Tipo_Accidente'] = ACCIDENTES.apply(Tipo_Accidente , axis=1)
ACCIDENTES['Choque_Nombre'] = ACCIDENTES.apply(Choque_Nombre , axis=1)
ACCIDENTES = ACCIDENTES[columns_to_consider]
ACCIDENTES.rename(columns={'TipoTiempo': 'TipoVia'}, inplace=True)
CONDUCTORES = diccionario_dataframes['CONDUCTORES']
VICTIMAS= diccionario_dataframes["VICTIMAS"]

ACCIDENTES.head()
```

```python
columnas = []
conteo_valores_unicos = []
valores_unicos = []
tipo_dato= []
conteo_nulos =[]
# Iterar a través de las columnas del DataFrame
for columna in ACCIDENTES.columns:
    valores_unicos_columna = ACCIDENTES[columna].unique()

    columnas.append(columna)
    conteo_valores_unicos.append(len(valores_unicos_columna))
    valores_unicos.append(valores_unicos_columna)
    tipo_dato.append(ACCIDENTES[columna].dtypes)
    conteo_nulos.append(ACCIDENTES[columna].isnull().sum())
informacion_columnas = pd.DataFrame({
    'Campo': columnas,
    'tipo_dato':tipo_dato,
    'Número_Valores_Unicos': conteo_valores_unicos,
    'conteo_nulos':conteo_nulos,
    'Valores_Unicos': valores_unicos
})

informacion_columnas.head(informacion_columnas.shape[0])
```

### Preprocesar los Datos de Accidentes y tratas vacios

```python
ACCIDENTES['Tipo_Accidente'] = ACCIDENTES['Tipo_Accidente'].str.replace(r' nan', '', regex=True)
ACCIDENTES['Choque_Nombre'] = ACCIDENTES['Choque_Nombre'].str.replace(r' nan', '', regex=True)
ACCIDENTES['Choque_Nombre'] = ACCIDENTES['Choque_Nombre'].replace('nan', 'No Aplica')
```

```python
columnas = []
conteo_valores_unicos = []
valores_unicos = []
tipo_dato= []
conteo_nulos =[]
# Iterar a través de las columnas del DataFrame
for columna in ACCIDENTES.columns:
    valores_unicos_columna = ACCIDENTES[columna].unique()

    columnas.append(columna)
    conteo_valores_unicos.append(len(valores_unicos_columna))
    valores_unicos.append(valores_unicos_columna)
    tipo_dato.append(ACCIDENTES[columna].dtypes)
    conteo_nulos.append(ACCIDENTES[columna].isnull().sum())
informacion_columnas = pd.DataFrame({
    'Campo': columnas,
    'tipo_dato':tipo_dato,
    'Número_Valores_Unicos': conteo_valores_unicos,
    'conteo_nulos':conteo_nulos,
    'Valores_Unicos': valores_unicos
})

informacion_columnas.head(informacion_columnas.shape[0])
```

```python
# Primero, crea un DataFrame para contar las víctimas por "idFormulario"
TOTAL_VICTIMAS = VICTIMAS.groupby('idFormulario').size().reset_index(name='TOTAL_VICTIMAS')
TOTAL_VICTIMAS
```

```python
# Obtiene los valores únicos en la columna "Peaton_Pasajero"
TIPOS_VICTIMAS = VICTIMAS['Peaton_Pasajero'].unique()

VICTIMAS_POR_TIPO = pd.DataFrame()

# Itera a través de los valores únicos en "Peaton_Pasajero" y crea una tabla dinámica para cada uno
for valor_unico in TIPOS_VICTIMAS:
    conteo_por_valor_unico = pd.pivot_table(VICTIMAS[VICTIMAS['Peaton_Pasajero'] == valor_unico],
                                           index='idFormulario',
                                           values='Peaton_Pasajero',
                                           aggfunc='count',
                                           fill_value=0)

    # Cambia el nombre de la columna a reflectir el valor único correspondiente
    conteo_por_valor_unico.columns = [valor_unico]

    # Combina los resultados en el DataFrame final
    VICTIMAS_POR_TIPO = pd.concat([VICTIMAS_POR_TIPO, conteo_por_valor_unico], axis=1)
    VICTIMAS_POR_TIPO.fillna(0, inplace=True)


VICTIMAS_POR_TIPO.reset_index(drop=False, inplace=True)
VICTIMAS_POR_TIPO.rename(columns={'Peaton': 'CantidadVictimasPeatones'}, inplace=True)
VICTIMAS_POR_TIPO.rename(columns={'Acompañante': 'CantidadVictimasAcompanante'}, inplace=True)
VICTIMAS_POR_TIPO.rename(columns={'Pasajero': 'CantidadVictimasPasajeros'}, inplace=True)
VICTIMAS_POR_TIPO
```

```python
TOTAL_CONDUCTORES = CONDUCTORES.groupby('idFormulario').size().reset_index(name='TOTAL_CONDUCTORES')
TOTAL_CONDUCTORES
```

```python
# Obtiene los valores únicos en la columna "VEHICULO_VIAJABA_CLASIFICADO"
TIPOS_CONDUCTORES = CONDUCTORES['VEHICULO_VIAJABA_CLASIFICADO'].unique()

CONDUCTORES_POR_TIPO = pd.DataFrame()

# Itera a través de los valores únicos en "VEHICULO_VIAJABA_CLASIFICADO" y crea una tabla dinámica para cada uno
for valor_unico in TIPOS_CONDUCTORES:
    conteo_por_valor_unico = pd.pivot_table(CONDUCTORES[CONDUCTORES['VEHICULO_VIAJABA_CLASIFICADO'] == valor_unico],
                                           index='idFormulario',
                                           values='VEHICULO_VIAJABA_CLASIFICADO',
                                           aggfunc='count',
                                           fill_value=0)

    # Cambia el nombre de la columna a reflectir el valor único correspondiente
    conteo_por_valor_unico.columns = [valor_unico]

    # Combina los resultados en el DataFrame final
    CONDUCTORES_POR_TIPO = pd.concat([CONDUCTORES_POR_TIPO, conteo_por_valor_unico], axis=1)
    CONDUCTORES_POR_TIPO.fillna(0, inplace=True)


CONDUCTORES_POR_TIPO.reset_index(drop=False, inplace=True)
CONDUCTORES_POR_TIPO['CantidadConductoresLivianos'] = CONDUCTORES_POR_TIPO['LIVIANO'] + CONDUCTORES_POR_TIPO['TAXI']
CONDUCTORES_POR_TIPO['CantidadConductoresPesados'] = CONDUCTORES_POR_TIPO['TRANSPORTE DE PASAJEROS'] + CONDUCTORES_POR_TIPO['CARGA']
CONDUCTORES_POR_TIPO.rename(columns={'MOTOCICLETA': 'CantidadConductoresMotocicleta'}, inplace=True)
CONDUCTORES_POR_TIPO.rename(columns={'BICICLETA': 'CantidadConductoresBICICLETA'}, inplace=True)

CONDUCTORES_POR_TIPO.drop(['SIN INFORMACION','TRANSPORTE DE PASAJEROS','LIVIANO','TAXI','CARGA'], axis=1, inplace=True)
CONDUCTORES_POR_TIPO
```

```python
# Obtiene los valores únicos en la columna "Sexo"
GENERO_CONDUCTORES = CONDUCTORES['Sexo'].unique()

GENERO_TIPOCONDUCTORES = pd.DataFrame()

# Itera a través de los valores únicos en "Peaton_Pasajero" y crea una tabla dinámica para cada uno
for valor_unico in GENERO_CONDUCTORES:
    conteo_por_valor_unico = pd.pivot_table(CONDUCTORES[CONDUCTORES['Sexo'] == valor_unico],
                                           index='idFormulario',
                                           values='Sexo',
                                           aggfunc='count',
                                           fill_value=0)

    # Cambia el nombre de la columna a reflectir el valor único correspondiente
    conteo_por_valor_unico.columns = [valor_unico]

    # Combina los resultados en el DataFrame final
    GENERO_TIPOCONDUCTORES = pd.concat([GENERO_TIPOCONDUCTORES, conteo_por_valor_unico], axis=1)
    GENERO_TIPOCONDUCTORES.fillna(0, inplace=True)


GENERO_TIPOCONDUCTORES.reset_index(drop=False, inplace=True)

GENERO_TIPOCONDUCTORES.rename(columns={'MASCULINO': 'CantidadConductoresHombres'}, inplace=True)
GENERO_TIPOCONDUCTORES.rename(columns={'FEMENINO': 'CantidadConductoresMujeres'}, inplace=True)

GENERO_TIPOCONDUCTORES.drop('NO APLICA', axis=1, inplace=True)
GENERO_TIPOCONDUCTORES
```

```python
MODELO_ANTIGUO = CONDUCTORES.groupby('idFormulario')['ModeloVehiculo'].min().reset_index()
modelo_vehiculo_replacement_value = int(np.mean(CONDUCTORES['ModeloVehiculo'].dropna().values))
MODELO_ANTIGUO
```

```python
ACCIDENTES_VF = ACCIDENTES.merge(TOTAL_VICTIMAS, on='idFormulario', how='left')
ACCIDENTES_VF = ACCIDENTES_VF.merge(VICTIMAS_POR_TIPO, on='idFormulario', how='left')
ACCIDENTES_VF = ACCIDENTES_VF.merge(TOTAL_CONDUCTORES, on='idFormulario', how='left')
ACCIDENTES_VF = ACCIDENTES_VF.merge(CONDUCTORES_POR_TIPO, on='idFormulario', how='left')
ACCIDENTES_VF = ACCIDENTES_VF.merge(GENERO_TIPOCONDUCTORES, on='idFormulario', how='left')
ACCIDENTES_VF = ACCIDENTES_VF.merge(MODELO_ANTIGUO, on='idFormulario', how='left')


if 'index' in ACCIDENTES_VF.columns:
    ACCIDENTES_VF.drop('index', axis=1, inplace=True)

# Rellena los valores NaN con 0 en las columnas especificadas
AJUSTAR_VACIOS = ['TOTAL_VICTIMAS', 'CantidadVictimasPeatones', 'CantidadVictimasAcompanante', 'CantidadVictimasPasajeros','TOTAL_CONDUCTORES','CantidadConductoresMotocicleta','CantidadConductoresBICICLETA','CantidadConductoresLivianos','CantidadConductoresPesados','CantidadConductoresHombres','CantidadConductoresMujeres']
ACCIDENTES_VF[AJUSTAR_VACIOS] = ACCIDENTES_VF[AJUSTAR_VACIOS].fillna(0)
ACCIDENTES_VF['ModeloVehiculo'] = ACCIDENTES_VF['ModeloVehiculo'].fillna(modelo_vehiculo_replacement_value).astype(int)
ACCIDENTES_VF
```

```python
columnas = []
conteo_valores_unicos = []
valores_unicos = []
tipo_dato= []
conteo_nulos =[]
# Iterar a través de las columnas del DataFrame
for columna in ACCIDENTES_VF.columns:
    valores_unicos_columna = ACCIDENTES_VF[columna].unique()

    columnas.append(columna)
    conteo_valores_unicos.append(len(valores_unicos_columna))
    valores_unicos.append(valores_unicos_columna)
    tipo_dato.append(ACCIDENTES_VF[columna].dtypes)
    conteo_nulos.append(ACCIDENTES_VF[columna].isnull().sum())
informacion_columnas = pd.DataFrame({
    'Campo': columnas,
    'tipo_dato':tipo_dato,
    'Número_Valores_Unicos': conteo_valores_unicos,
    'conteo_nulos':conteo_nulos,
    'Valores_Unicos': valores_unicos
})

informacion_columnas.head(informacion_columnas.shape[0])
```
